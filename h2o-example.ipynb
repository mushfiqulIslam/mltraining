```python
# https://seanzhang-data.medium.com/h2o-automl-in-python-comprehensive-tutorial-f25001c11b80
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init(max_mem_size='8G')
```

```python
train = h2o.import_file("train.csv")
test = h2o.import_file("test.csv")
```

```python
x = train.columns
y = "Survived"
train[y] = train[y].asfactor()
x.remove(y)
```

```python
%%time
aml = H2OAutoML(max_models=20, max_runtime_secs=12000)
aml.train(x=x, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head(rows=15)
```

```python
aml.leader # Provides Best model
```

```python
aml.leader.varimp()
model = h2o.get_model("StackedEnsemble_BestOfFamily_1_AutoML_1_20240130_203419")
```
